```python
#! pip install pandas e matplotlib

import pandas as pd
import matplotlib.pyplot as plt
```

```python
FNAME = 'Data_Intro_Pandas.xlsx'

df = pd.read_excel(f'{FNAME}')
print(df)
```

```python
print(df.head(2))
```

```python
# Determinar mínimo
min = df.min()
print(min)
```

```python
# Determinar máximo
max = df.max()
print(max)
```

```python
# Determinar média
med = df.mean()
print(med)
```

```python
# Determinar desvio padrão
desv = df.std()
print(desv)
```

```python
# Representação gráfica
df.plot()
```

```python
# Figura
df.plot.scatter(x="Peso", y="Idade")

plt.xlabel('Peso (Kg)')
plt.ylabel('Idade (anos)')

plt.savefig('Scatter.png')
```
